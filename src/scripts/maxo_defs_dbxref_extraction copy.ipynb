```python
#this is a python notebook to identify definitions in MAXO which have been extracted from other ontologies. This is step 1 in setting up a pattern to automatically update those definitions from outside ontologies.
```

```python
import re

maxo = open('maxo-edit.txt')
maxoid=list()
dbxref=list()
ontoxrefcount=0
for line in maxo: 
    line =line.strip()
    line=line.strip('/')
    #There are 2 types of hasDBXrefs, one that is an xref to the MAXO term, and the other is an xref to the definion. The 'startswith' line identifies only the xrefs associated with definitions
    if line.startswith('AnnotationAssertion(Annotation(<http://www.geneontology.org/formats/oboInOwl#hasDbXref>') :
        #x finds all things that sorta look ontology xrefs in the line. Some other things that look like xrefs are 'rdfs:comment', 'http:'xrefs, 'isbn-10'xrefs.
        x=re.findall('[A-Z]{4,11}:[A-Z0-9]{1,15}', line)
        for item in x:
            # y is a list without empty lines
            if len(item)>2:
                y=x
                #this finds the MAXO ID associated with the xref. We use 'line' because this would not be within the x.findall
                maxoid=re.findall('MAXO_[0-9]*', line)
                for unit in y:
                    #this removes *most* PMIDs...
                    if not unit.startswith('PMID'):
                        #this removes any definition with more than 1 xref (some of these have PMIDs that weren't eliminated) because if there are 2 xrefs listed, it is unclear which definition should be used.
                        if len(y)==1:
                            xref=y
                            ontoxrefcount=ontoxrefcount+1
                            print(maxoid, xref)
        
print('xrefs', ontoxrefcount)
```
